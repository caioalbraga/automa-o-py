```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import pyautogui as py
import time


# Resposta fixa

resposta_1 = "Resposta 1: a) Unidos pelo mesmo Espírito e irão entender que Deus trata com Israel como trata com a igreja. b) Compartilhando do mesmo alimento. c) Conversaríamos usando a mesma linguagem, a mesma comunhão."
resposta_2 = "Resposta 2: Quem está falando é a igreja fiel que foi arrebatada. a) Israel ensinaria os mistérios do Velho Testamento.  b) Batismo com o Espírito Santo. c) Ensinaria a Israel o que é o Pentecostes. O Vinho da cura."
resposta_3 = ""
resposta_4 = ""
resposta_fixa = f"{resposta_1}\n\n{resposta_2}\n\n{resposta_3}\n\n{resposta_4}"

# Inicializar o driver do Selenium (certifique-se de ter o driver do navegador correto instalado)
navegador = webdriver.Chrome()

# Definir um tempo de espera máximo para todos os elementos
navegador.implicitly_wait(10)  # Isso significa que o Selenium esperará até 10 segundos por cada operação de localização de elementos

# Navegar para a página
navegador.get('https://www.igrejacristamaranata.org.br/ebd/participacoes/')

# Ler o arquivo CSV com o Pandas
dados = pd.read_csv('rodrigues.csv', dtype= str)

try:
    # Aguardar até que o campo CPF esteja visível
    WebDriverWait(navegador, 10).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="icmEbdNacionalForm"]/div[2]/div/div[1]/input')))
    
    # Iterar sobre as linhas do DataFrame
    for indice, linha in dados.iterrows():
        #ebd N (?)
        WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="icmEbdNacionalForm"]/div[1]/div/div/div/input'))).click()
        time.sleep(5)
        # Obter os valores de cada coluna
        valor_cpf = str(linha['CPF'])  # Convertendo para string
        
        # Localizar e preencher os campos na página da web
        campo_cpf = navegador.find_element(By.XPATH, '//*[@id="icmEbdNacionalForm"]/div[2]/div/div[1]/input')
        campo_cpf.clear()
        campo_cpf.send_keys(valor_cpf)
        
        # Pausa de 2 segundos antes de continuar
        time.sleep(1)
        
        py.scroll(-1300)
        
        # Preencher a resposta fixa
        campo_resposta = navegador.find_element(By.XPATH, '//*[@id="icmEbdNacionalForm"]/div[5]/div/div[2]/div/div[2]/div[1]')
        campo_resposta.clear()
        campo_resposta.send_keys(resposta_fixa)
        
        # Pausa de 2 segundos antes de continuar
        time.sleep(1)
        
        # Aguardar até que o botão de aceitar os termos esteja presente
        WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="icmEbdNacionalForm"]/div[6]/div/div/div/label'))).click()

        # Agurdar até o botão categoria esteja presente         
        WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="icmEbdNacionalForm"]/div[5]/div/div[1]/input[2]'))).click()

        # Enviar participação 
        
        WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="icmEbdNacionalForm"]/button'))).click()

        # Pausa de 2 segundos antes de continuar
        time.sleep(2)
        WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[4]/div/div[4]/div/button'))).click()
        
        py.scroll(5000)
        
        
        # Clicar em um botão ou executar qualquer outra ação necessária após preencher os campos
        
except Exception as e:
    print("Erro ao preencher formulário:", e)




```
